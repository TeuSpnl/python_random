```python
"""
## 1. Faturamento do Melhor e do Pior Mês do Ano

Qual foi o valor de vendas do melhor mês do Ano?
E valor do pior mês do ano?
"""

meses = ['janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho', 'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro']
vendas_1sem = [25000, 29000, 22200, 17750, 15870, 19900]
vendas_2sem = [19850, 20120, 17540, 15555, 49051, 9650]

vendas_total = vendas_1sem + vendas_2sem
print(vendas_total)

maior = max(vendas_total)
menor = min(vendas_total)

print(f"max: {maior}, min: {menor}")

pos_max = vendas_total.index(maior)
pos_min = vendas_total.index(menor)

mes_max = meses[pos_max]
mes_min = meses[pos_min]
```

```python
print(f"O mês que teve maior valor de vendas foi {mes_max} e o que teve menor valor de vendas foi {mes_min}.")

soma = sum(vendas_total)
repres_maior = maior / soma

print(f"O faturamento total foi R${soma:.2f} e o melhor mês representa {repres_maior:.2%}% do total.")
```
